### Amigo secreto
- Una app que pida nombres para jugar amigo secreto
- Va a pedir nombres hasta que reciba un valor vacío (un enter) un false
- Va pintar el amigo secreto de cada uno con la condición de que un amigo secreto no debe ser uno mismo (DOS LISTAS)

```python
import random

def names():
    name_list = []
    
    while True:
        nombre = input("Introduce un nombre (Enter para terminar): ")
        if nombre == "":
            break
        name_list.append(nombre)

    friend_dict = {}

    for x in name_list:
        amigo = random.choice(name_list)
        while amigo == x:
            amigo = random.choice(name_list)
        friend_dict[x] = amigo

    return friend_dict

names()

    
```

### Contador de palabras mejorado
Crea una función que reciba un string y devuelva un diccionario donde:

- La clave sea cada palabra.
- El valor sea el número de veces que aparece.
- Ignore mayúsculas y minúsculas.
- No tenga en cuenta puntos ni comas.

```python
# VERSIÓN CON COUNT()

entry_phrase = str(input("Escribe una frase: "))

def dict_words(phrase):
    phrase_lower = phrase.lower()
    phrase_replace = phrase_lower.replace(",", "").replace(".", "")
    phrase_list = phrase_replace.split() # Para devolver palabras en modo lista
    words_dict = {}
    for x in phrase_list:
        words_dict[x] = phrase_list.count(x)
    return words_dict
print(f"Este es el diccionario con las palabras y las cantidades de veces que se repiten: {dict_words(entry_phrase)}")
```

### Diccionario de estudiantes
Crea una función donde el diccionario tiene esta estructura:
{
    "Ana": [7, 8, 9],
    "Luis": [5, 6, 7],
    "Marta": [10, 9, 8]
}

La función debe:
- Calcular el promedio de cada estudiante.
- Devolver un nuevo diccionario con:
- Nombre como clave
- Promedio como valor

```python
student_dict = {
    "Ana": [7, 8, 9],
    "Luis": [5, 6, 7],
    "Marta": [10, 9, 8]
}

def avg_student(students):
    new_dict = {}
    for x in students:
        value = students[x] # Acceder a ítems[] Obtengo los values de cada estudiante en una variable (value<classList>) independiente para c/u
        avg_value = sum(value)/len(value) # Calculo el AVG de cada uno
        new_dict[x] = avg_value
    return new_dict
print(avg_student(student_dict))

```

```python
# Versión mejorada con .items() para calcular dos valores:
student_dict = {
    "Ana": [7, 8, 9],
    "Luis": [5, 6, 7],
    "Marta": [10, 9, 8]
}

def avg_student(students):
    new_dict = {}
    for key, value in students.items(): # items permite calcular dos valores simultaneamente y trabajar con ellos por separado
        avg_value = sum(value) / len(value) # como value es de clase lista se suman los valores y se dividen para obtener el promedio
        new_dict[key] = avg_value # se asocian las key de siempre al cálculo del promedio
    return new_dict
print(avg_student(student_dict))
```

### Análisis de ventas
Tienes esta lista de diccionarios:

ventas = [
    {"producto": "Laptop", "precio": 1000, "cantidad": 2},
    {"producto": "Mouse", "precio": 50, "cantidad": 5},
    {"producto": "Teclado", "precio": 80, "cantidad": 3}
]

Crea una función que:

- Calcule el total vendido por producto (precio * cantidad)
- Devuelva un diccionario con: 
- Producto como clave
- Total generado como valor

```python
ventas = [
    {"producto": "Laptop", "precio": 1000, "cantidad": 2},
    {"producto": "Mouse", "precio": 50, "cantidad": 5},
    {"producto": "Teclado", "precio": 80, "cantidad": 3}
]

def total_por_producto(lista_ventas):
    resultado = {}

    for dict_indiv in lista_ventas: # venta es cada diccionario individual
        # primero se desagrupa el diccionario por los values
        producto = dict_indiv["producto"] # accediendo a los value: laptop, mouse, teclado
        precio = dict_indiv["precio"] # accediendo a los value: 1000, 50, 80
        cantidad = dict_indiv["cantidad"] # aaccediendo a los value: 2, 5, 3

        total = precio * cantidad 

        resultado[producto] = total # añadir al diccionario llamado resultado, el key de producto y el total

    return resultado


print(total_por_producto(ventas))
```

### Lista números mayor, menor y promedio

Crea una función que reciba una lista [12, 45, 2, 19, 33, 67, 21, 5]. La función debe recorrer la lista con un for y encontrar:
- número mayor
- número menor
- promedio


La función debe devolver estos tres elementos sin usar max(), min() ni sum()


```python
lista_numeros = [12, 45, 2, 19, 33, 67, 21, 5]

def numerica(numeros):
    num_menor = numeros[0]
    num_mayor = numeros[0]
    total_suma = 0
    for x in numeros:
        total_suma = total_suma + x
        if x > num_mayor: 
            num_mayor = x 
        if x < num_menor:
            num_menor = x
    promedio = total_suma/len(numeros)
    # return promedio, num_mayor, num_menor
    return {
        "número mayor": num_mayor,
        "número menor": num_menor,
        "promedio": promedio
}
print(numerica(lista_numeros))
```

Dados los siguientes diccionarios:

shop1 = {"leggins": 10, "jeans": 5, "vestidos": 7}
shop2 = {"leggins": 3, "abrigos": 8, "vestidos": 2}
shop3 = {"calcetines": 2, "bragas": 30, "vestidos": 1}

Crea una función que devuelva un nuevo diccionario con:
- Productos de los tres diccionarios
- Si el producto está en los tres, se deben sumar las cantidades

```python
shop1 = {"leggins": 10, "jeans": 5, "vestidos": 7}
shop2 = {"leggins": 3, "abrigos": 8, "vestidos": 2}
shop3 = {"calcetines": 2, "bragas": 30, "vestidos": 1}

def shops(tienda1, tienda2, tienda3):
    shop4 = tienda1.copy()
    for x in tienda2:
        if x in shop4:
            shop4[x] = tienda2[x] + shop4[x]
        else:
            shop4[x] = tienda2[x]
    for x in tienda3:
        if x in shop4:
            shop4[x] = tienda3[x] + shop4[x]
        else:
            shop4[x] = tienda3[x]
    return shop4
print(shops(shop1, shop2, shop3))
```

### Separar pares e impares

Crea una función que reciba la siguiente entrada: [3,6,8,1,9,2]. La función debe devolver dos listas: una de números pares y otra de números impares.



```python
entrada_numeros = [3,6,8,1,9,2]

def listados_par_impar(listado):
    lista_par = []
    lista_impar = []
    for x in listado:
        if x % 2 == 0: # si fuera impar: if numero % 2 != 0:
            lista_par.append(x)
        else:
            lista_impar.append(x)
    return f"Estos son los números pares: {lista_par}", f"Estos son los números impares: {lista_impar}"

listados_par_impar(entrada_numeros)
```

### BONUS: Aprobados, suspendidos, promedio

Dada la siguiente lista: 
students = [
{"name":"Ana","grade":7},
{"name":"Luis","grade":4},
{"name":"Marta","grade":9},
{"name":"Carlos","grade":5}
]

Crea una función que devuelva un diccionario con listas de los aprobados, los suspendidos y el promedio de notas de todos los estudiantes.

```python
students = [
{"name":"Ana","grade":7},
{"name":"Luis","grade":4},
{"name":"Marta","grade":9},
{"name":"Carlos","grade":5}
]


def analizar_estudiantes(lista_estudiantes):

    aprobados = []
    suspendidos = []
    suma_notas = 0

    for student in lista_estudiantes:
        nombre = student["name"]
        nota = student["grade"]

        suma_notas += nota

        if nota >= 6.5:
            aprobados.append(nombre)
        else:
            suspendidos.append(nombre)

    promedio = suma_notas / len(lista_estudiantes)

    return {
        "aprobados": aprobados,
        "suspendidos": suspendidos,
        "promedio": promedio
    }


resultado = analizar_estudiantes(students)
print(resultado)
```
