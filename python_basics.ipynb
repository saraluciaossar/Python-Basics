# M7T8
Crea una funció que:
* Demani a l’usuari 3 noms de ciutats, un per un.
* Comprovi si cada ciutat conté la lletra 'E' o 'e' i té més de 5 lletres.
* Si la ciutat compleix aquestes condicions, afegeix-la a una llista.
* Quan hagi acabat, retorna la llista completa de ciutats que compleixen els requisits.
* Finalment, crida la funció i imprimeix la llista completa.

```python
# Versión chambona

print("""Bienvenido/a,
A continuación, vas a escribir tres nombres de ciudades diferentes: """)

def three_cities():
    city_1 = input("Introduce el primer nombre: ")
    city_2 = input("Introduce el segundo nombre: ")
    city_3 = input("Introduce el tercer nombre: ")

    cities_list = []

    if len(city_1) > 5 and ("E" in city_1 or "e" in city_1):
        cities_list.append(city_1)
    if len(city_2) > 5 and ("E" in city_2 or "e" in city_2):
        cities_list.append(city_2)
    if len(city_3) > 5 and ("E" in city_3 or "e" in city_3):
        cities_list.append(city_3)

    print(f"Esta(s) ciudad(es) tiene(n) la letra 'e': {cities_list}")

three_cities()
```

# M7T10
1. Crea una funció que rebi una llista de nombres i retorni una nova llista amb només els nombres parells.
2. Després, crea una llista amb 10 números, i:
* Imprimeix quants nombres parells hi ha.
* Imprimeix quins són aquests nombres parells.


```python
list_1 = [10, 15, 20, 25, 30, 35, 40, 45]

def solo_pares(list_funct):
    par_list = []
    for x in list_funct:
      if x % 2 == 0:
        par_list.append(x)
    return par_list #EN EL RETURN SE ACABA LA FUNCIÓN
print(solo_pares(list_1)) #PARA PODER VER QUÉ FUE LO QUE SE HIZO EN LA FUNCIÓN, LA IMPRIMO


# Després, crea una llista amb 10 números, i:

list_2 = [3, 6, 9, 12, 15, 18, 21, 24, 27, 30]
print(len(solo_pares(list_2))) # Imprimeix quants nombres parells hi ha.
print(solo_pares(list_2)) # Imprimeix quins són aquests nombres parells.
```

# M7T11
Escriu un programa que faci el següent:
- Demani a l’usuari que introdueixi una frase.
- Utilitza una funció que rebi aquesta frase com a argument i retorni un diccionari amb:
- Les paraules que apareixen a la frase com a claus.
- El nombre de vegades que apareixen com a valors.
- No utilitzis la funció integrada count() per comptar les paraules.
Finalment, mostra el diccionari per pantalla.

### is none

```python
user_entry = input("Introduce una frase: ")

def dict_phrases(dict_entry):
    slicing_words = list(dict_entry.split())
    dict_1 = {}
    for x in slicing_words:
        value = dict_1.get(x)
        if value is None:
            dict_1[x] = 1
        else:
             dict_1[x] = value + 1
    print(dict_1)
dict_phrases(user_entry)
```

### if in

```python
user_entry = input("Introduce una frase: ")

def dict_phrases(dict_entry):
    slicing_words = list(dict_entry.split())
    dict_1 = {}
    for x in slicing_words:
        if x in dict_1:
            dict_1[x] += 1
        else:
             dict_1[x] = 1
    print(dict_1)
dict_phrases(user_entry)
```
