### Ejercicio 2 - condicionales if, def, .loc y lambda

Crea una nueva columna llamada popularitat con estas reglas:

num_subscribers > 50000 → "Alta"

10000 <= num_subscribers <= 50000 → "Mitjana"

num_subscribers < 10000 → "Baixa"

```python
# Con .loc
df_udemy_ex2 = df_udemy.copy()
df_udemy_ex2["popularitat"] = ""

df_udemy_ex2.loc[df_udemy_ex2["num_subscribers"] > 50000, "popularitat"] = "Alta"
# df_udemy_ex2.loc[(df_udemy_ex2["num_subscribers"] >= 10000) & (df_udemy_ex2["num_subscribers"] <= 50000), "popularitat"] = "Mitjana"
df_udemy_ex2.loc[df_udemy_ex2["num_subscribers"].between(10000, 50000), "popularitat"] = "Mitjana"
df_udemy_ex2.loc[df_udemy_ex2["num_subscribers"] < 10000, "popularitat"] = "Baixa"

df_udemy_ex2.tail(10)
```

```python
# Con .loc EXPLICACIÓN PASO A PASO

# todas las filas quedan con texto vacío.
df_udemy_ex2["popularitat"] = ""

df_udemy_ex2.loc[df_udemy_ex2["num_subscribers"] > 50000, "popularitat"] = "Alta"
# .loc sirve para seleccionar filas y columnas.
# Aquí estamos diciendo:
# en las filas donde "num_subscribers" sea mayor que 50000,
# en la columna "popularitat", escribe "Alta".

df_udemy_ex2.loc[(df_udemy_ex2["num_subscribers"] >= 10000) & (df_udemy_ex2["num_subscribers"] <= 50000),"popularitat"] = "Mitjana"
# Aquí seleccionamos las filas donde "num_subscribers" esté entre 10000 y 50000.
# Ojo:
# - >= 10000 significa "mayor o igual que 10000"
# - <= 50000 significa "menor o igual que 50000"
# - & significa "y", o sea, las dos condiciones deben cumplirse a la vez.
# En esas filas, en la columna "popularitat", se escribe "Mitjana".

df_udemy_ex2.loc[df_udemy_ex2["num_subscribers"] < 10000, "popularitat"] = "Baixa"
# Aquí seleccionamos las filas donde "num_subscribers" sea menor que 10000.
# En esas filas, en la columna "popularitat", se escribe "Baixa".
```

```python
# Con def
def popularidad(x):
    if x > 50000:
        return "Alta"
    elif 10000 <= x <= 50000:
        return "Mitjana"
    else:
        return "Baixa"

df_udemy_ex2["popularitat"] = df_udemy_ex2["num_subscribers"].apply(popularidad)

df_udemy_ex2.head()

# Definimos una función llamada "popularidad"
# x representa cada valor individual de la columna num_subscribers
# Si el valor es mayor que 50000, devuelve "Alta"
# Si el valor está entre 10000 y 50000, devuelve "Mitjana"
# # Si no se cumple ninguna de las condiciones anteriores
# # significa que es menor que 10000, así que devuelve "Baixa"
# apply aplica la función "popularidad" a cada valor X de la columna "num_subscribers". Apply es quien elige a qué aplicarle la función
# El resultado de cada fila se guarda en la nueva columna "popularitat"
```

```python
df_udemy_ex2["popularitat"] = df_udemy_ex2["num_subscribers"].apply(
    lambda x: "Alta" if x > 50000
    else "Mitjana" if 10000 <= x <= 50000
    else "Baixa"
)

df_udemy_ex2.head()
# apply recorre cada valor de la columna "num_subscribers".
# lambda x crea una función pequeña donde x representa cada valor de la columna.
# Si x > 50000 devuelve "Alta".
# Si no, revisa la siguiente condición: si x está entre 10000 y 50000 devuelve "Mitjana".
# Si no se cumple ninguna de las anteriores, devuelve "Baixa".
# El resultado final se guarda en la nueva columna "popularitat".
```

### Ejercicio 2 - retorna un diccionario por subgrupos (.get_group())

Crea una función llamada course_level que reciba un DataFrame y un instructional_level y filtre solo ese nivel. Calcule:
- cantidad de cursos
- media de rating
- total de suscriptores
- retorne esos resultados

```python
df_udemy = pd.read_csv("udemy_courses.csv")

level_1 = "Beginner Level"
level_2 = "Intermediate Level"
level_3 = "Expert Level"

# SOLUCIÓN CON .get_group()
def course_level(df, level):

    df_level = df.groupby("instructional_level").get_group(level)

    total_number = df_level["id"].count()
    avg_course = df_level["rating"].mean()
    total_subscribers = df_level["num_subscribers"].sum()

    return (
        f"El total de cursos en este nivel es de: {total_number} cursos",
        f"El promedio de rating para este nivel es: {avg_course}",
        f"El total de suscriptores para este nivel es de: {total_subscribers}"
    )

course_level(df_udemy, level_3)


# SOLUCIÓN CON .get_group() y agg({})
def course_level(df, level):
    result = df.groupby("instructional_level").get_group(level).agg({
        "id": "count",
        "rating": "mean",
        "num_subscribers": "sum"
    })
    return (
        f"El total de cursos en este nivel es de: {int(result['id'])} cursos", # Le puse int porque el resultado me salía como float y realmente es un número entero
        f"El promedio de rating para este nivel es: {result['rating']}",
        f"El total de suscriptores para este nivel es de: {int(result['num_subscribers'])}"
    )
course_level(df_udemy, level_3)
# df_udemy.head()


# resumen = df.groupby('categoria').agg(
#     total_ventas=('ventas', 'sum'),
#     numero_registros=('ventas', 'count')

# )
 
```

### Ejercicio 4 - retorna con filtrado por condicionales

La función debe:
- Crear una nueva columna llamada popularity_type.
- Clasificar los cursos según el número de suscriptores (num_subscribers) usando el límite recibido en la función.

Por ejemplo, si el límite es 50 000:

num_subscribers > limit → "Very popular"

10000 <= num_subscribers <= limit → "Moderately popular"

num_subscribers < 10000 → "Low popularity"

```python
# CON .LOC
df_udemy = pd.read_csv("udemy_courses.csv")
user_limit_popularity = 50000

def price_limit(df, limit):

    df_popularity = df.copy()
    df_popularity["popularity"] = ""
    df_popularity.loc[df_popularity["num_subscribers"] > limit, "popularity"] = "Very popular"
    df_popularity.loc[(df_popularity["num_subscribers"] >= 10000) & (df_popularity["num_subscribers"] <= limit), "popularity"] = "Moderately popular"
    df_popularity.loc[df_popularity["num_subscribers"] < 10000, "popularity"] = "Low popularity"

    return df_popularity

price_limit(df_udemy, user_limit_popularity)
```

```python
# CON DOBLE DEF
df_udemy = pd.read_csv("udemy_courses.csv")
user_limit_popularity = 50000

def price_limit(df, limit):    

    df_popularity = df.copy()
    
    def popularity(x):
        if x > limit:
            return "Very popular"
        elif 10000 <= x <= limit:
            return "Moderately popular"
        else:
            return "Low popularity"
    
    df_popularity["Popularity"] = df_popularity["num_subscribers"].apply(popularity)
    return df_popularity

price_limit(df_udemy, user_limit_popularity)
```

### Ejercicio 6 - función que recibe df y número, crea columna calculada y retorna top nlargest

Crea una función llamada top_cursos(df, n) que reciba: un DataFrame y un número

La función debe:
- crear una nueva columna llamada index_rendiment con esta fórmula: (num_subscribers * rating) / num_reviews
- crear una nueva columna llamada TipusCurs: ""Premium" si rating >= 4.5 y num_subscribers >= 50000, "Standard" si rating >= 4.0 y rating < 4.5 y num_subscribers >= 10000, "Basic" en cualquier otro caso
- Ordenar el DataFrame de mayor a menor por index_rendiment	
- Retornar solo los n primeros cursos mostrando: title, rating, num_reviews, num_subscribers, TipusCurs

```python
df_udemy = pd.read_csv("udemy_courses.csv")
number = 10

def top_cursos(df, n):
    df_copy = df.copy()

    df_copy["index_rendiment"] = df_copy["num_subscribers"] * df_copy["rating"] / df_copy["num_reviews"]

    df_copy["TipusCurs"] = "Basic"
    
    df_copy.loc[(df_copy["rating"] >= 4.5) & (df_copy["num_subscribers"] >= 50000), "TipusCurs"] = "Premium"
    df_copy.loc[(df_copy["rating"] >= 4.0) & (df_copy["rating"] < 4.5) & (df_copy["num_subscribers"] >= 10000), "TipusCurs"] = "Standard"

    return df_copy[["title","rating","num_reviews","num_subscribers","TipusCurs","index_rendiment"]].nlargest(n, "index_rendiment")
top_cursos(df_udemy, number)

```

### Ejercicio 8 - ~isin() isnot()
Crea una función llamada hospitales_no_catalunya(df) que:

- filtre los hospitales que NO estén en "Cataluña" ni "Madrid"
- entre esos hospitales, seleccione solo los que tengan más de 200 camas y retorne únicamente las columnas: hospital_id, comunidad_autonoma, num_camas, porcentaje_ocupacion

```python
data = pd.read_csv("hospitales.csv", sep=";", encoding="latin-1")

def hospitales_no_catalunya(df):

    df_filtrado = df.loc[
        (~df["comunidad_autonoma"].isin(["Cataluña", "Madrid"])) &
        (df["num_camas"] > 200),
        ["hospital_id", "comunidad_autonoma", "num_camas", "porcentaje_ocupacion"]
    ]

    return df_filtrado

hospitales_no_catalunya(data)
```
